```python
import matplotlib.pyplot as plt

# List of columns to plot
columns = ['dne', 'srl', 'srrn', 'eda']

# Define the figure and set the number of subplots
fig, axes = plt.subplots(len(columns), 1, figsize=(10, 12), sharex=True)

# Plot each column in a separate subplot
for i, column in enumerate(columns):
    axes[i].plot(data['time'], data[column], label=column, color='tab:blue')
    axes[i].set_title(f'{column.upper()} Over Time')
    axes[i].set_ylabel(column.upper())
    axes[i].legend()

# Set a common x-axis label
plt.xlabel('Time')

# Adjust layout
plt.tight_layout()  

# Show the plot
plt.show()
```
